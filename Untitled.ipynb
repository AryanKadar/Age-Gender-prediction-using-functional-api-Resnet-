```python
import pandas as pd
import os
import cv2
import matplotlib.pyplot as mp
import numpy as np
import random
```

```python
import tensorflow as tf
import pickle
from tensorflow import keras 
from keras import Sequential
from keras.layers import *
from tensorflow.keras.callbacks import EarlyStopping as ES
from tensorflow.keras.applications import ResNet50,VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator as IDG
from keras.models import Model
haar=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
from tensorflow_addons.metrics import RSquare as r2
```

```python
data=[]
path_=[]
path=os.path.join("archive","UTKFace")
for file in os.listdir(path):
    img_path=os.path.join(path,file)
    path_.append(img_path.split("\\")[2])
    img=cv2.imread(img_path)
    image=cv2.resize(img,(224,224))
    data.append(image)
    
```

```python
Gender=[]
Age=[]
for i in path_:
    if int(i.split("_")[1])==0:
        Gender.append(0)
    else:
        Gender.append(1)
    Age.append(int(i.split("_")[0].split("\\")[-1]))
```

```python
t=pd.DataFrame()
t["Age"]=Age
t["Gender"]=Gender
t["Path"]=path_
```

```python
t = t.sample(frac = 1)
t.head()
```

```python
Res=ResNet50(
    weights="imagenet",
    include_top=False,
    input_shape=(224,224,3)
    
)
Res.trainable=False
```

```python
for i in Res.layers:
    print(i.name,"=",i.trainable)
```

```python
for i in Res.layers:
    if "conv5_block3" in i.name:
        i.trainable=True
```

```python
for i in Res.layers:
    print(i.name,"=",i.trainable)
```

```python
Res.summary()
```

```python
input=Res.layers[-1].output
f=Flatten()(input)
h1___=Dense(20,activation="relu")(f)
h1__=Dense(30,activation="relu")(h1___)
h1_=Dense(40,activation="relu")(h1__)
h1=Dense(90,activation="relu")(h1_)
o1=Dense(1,activation="linear",name="Age")(h1)

h2_=Dense(50,activation="relu")(f)
h2=Dense(20,activation="relu")(h2_)
o2=Dense(1,activation="sigmoid",name="Gender")(h2)
```

```python
t.head()
```

```python
model=Model(inputs=Res.input,outputs=[o1,o2])
```

```python
model.summary()
```

```python
datagen = IDG(featurewise_center=True,brightness_range=(0.6, 1),fill_mode="constant",rotation_range=20,width_shift_range=0.2,height_shift_range=0.2,horizontal_flip=True,validation_split=0.2)
```

```python
es=ES(monitor="val_loss",
    min_delta=0.00001,
    patience=4,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
    start_from_epoch=0,)
```

```python
train=datagen.flow_from_dataframe(t,batch_size=8,directory=path,x_col="Path",y_col=["Age","Gender"],target_size=(224,224),class_mode="multi_output",subset='training')
test=datagen.flow_from_dataframe(t,batch_size=8,directory=path,x_col="Path",y_col=["Age","Gender"],target_size=(224,224),class_mode="multi_output",subset='validation')
```

```python
model.compile(optimizer="adamax",loss={"Age":tf.keras.losses.mae,"Gender":"binary_crossentropy"},metrics={"Age":r2(),"Gender":"accuracy"},loss_weights={"Age":0.01,"Gender":0.10})
```

```python
history=model.fit(train,batch_size=40,steps_per_epoch=(len(t)*0.8)//8,validation_data=test,validation_steps=(len(t)*0.2)//8,epochs=3,callbacks=es)
```

```python
#model.save('face_A_G_model')
#pickle.dump(datagen, open("face_A_G_.pkl", 'wb'))
model=keras.models.load_model('face_A_G_model')
```

```python
cap=cv2.VideoCapture(0)
data_aryan=[]
while True:
    ret,frame=cap.read()
    frame1 = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    co_ord=haar.detectMultiScale(frame1)
    max_co_ord=[]
    index_co_ord=None
    for i,j,w,h in co_ord:
        if co_ord!=():
            max_co_ord.append(abs((w+i)-i)+abs((h+j)-j))
    if co_ord!=() and max(max_co_ord)>160:
        index_co_ord=max_co_ord.index(max(max_co_ord))
        x_=co_ord[index_co_ord][0]
        y_=co_ord[index_co_ord][1]
        w_=co_ord[index_co_ord][2]
        h_=co_ord[index_co_ord][3]
        frame_1=frame[y_:h_+y_,x_:x_+w_]
        frame_1=cv2.resize(frame_1,(224,224))
        kl=model.predict(frame_1.reshape(1,224,224,3))
        if kl[1][0][0]<0.5:
            cv2.rectangle(frame,(x_,y_),(x_+w_,y_+h_),(0,255,0),1)
            cv2.putText(frame,"Male Age-"+str(kl[0][0][0]),(x_,y_),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2,cv2.LINE_AA)
        else:
            cv2.rectangle(frame,(x_,y_),(x_+w_,y_+h_),(0,255,0),1)
            cv2.putText(frame,"Female Age-"+str(kl[0][0][0]),(x_,y_),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2,cv2.LINE_AA)
            
    else:
        cv2.putText(frame,"Face Not Found",(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,255),2,cv2.LINE_AA)
    cv2.imshow("window",frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
